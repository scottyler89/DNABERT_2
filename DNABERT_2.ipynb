```python
import torch
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
model = AutoModel.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
```

```python
print(torch.cuda.is_available())
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

print(model)

dna = "ACGTAGCATCGGATCTATCTATCGACACTTGGTTATCGATCTACGAGCATCTCGTTAGCACGTAGCATCGGATCTATCTATCGACACTTGGTTATCGATCTACGAGCATCTCGTTAGC"
inputs = tokenizer(dna, return_tensors = 'pt')["input_ids"].to(device)
hidden_states = model(inputs)[0] # [1, sequence_length, 768]
print(hidden_states.shape)

dna = "ACGTAGCATCGGATCTATCTATCGACACTTGGTTATCGATCTACGAGCATCTCGTTAGC"
inputs = tokenizer(dna, return_tensors = 'pt')["input_ids"].to(device)
hidden_states = model(inputs)[0] # [1, sequence_length, 768]
print(hidden_states.shape)

# embedding with mean pooling
#embedding_mean = torch.mean(hidden_states[0], dim=0)
#print(embedding_mean.shape) # expect to be 768

# embedding with max pooling
#embedding_max = torch.max(hidden_states[0], dim=0)[0]
#print(embedding_max.shape) # expect to be 768

```

```python
tokenization_output = tokenizer.encode_plus(dna, return_tensors='pt', add_special_tokens=True, return_offsets_mapping=True)
input_ids = tokenization_output["input_ids"]
offsets = tokenization_output["offset_mapping"] # This gives you the character offsets for each token
tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
for token, offset in zip(tokens, offsets[0]):
    original_substring = dna[offset[0]:offset[1]]
    print(f"Token: {token} | Original Substring: {original_substring}")
```
